```python
# checkpoint results
import pickle
```

```python
def load_dataset(filename):
    
    rename_dict = {
        'aff' : 'Affiliates',
        'auddig' : 'Digital Audio',
        'audtr' : 'Radio',
        'dm' : 'Direct Mail',
        'em' : 'Email',
        'inst' : 'Insert',
        'nsp' : 'Newspaper',
        'on' : 'Online Display',
        'sem' : 'SEM',
        'sms' : 'SMS',
        'so' : 'Social',
        'viddig' : 'Digital Video',
        'vidtr' : 'TV',
        'y_baseline' : 'Baseline Sales',
    }
    
    with open(filename, 'rb') as f:
        dataset= pickle.load(f)
    
    perc_contrib, perc_incr_contrib = dataset.get_chn_percent_contrib()
    perc_incr_contrib = pd.DataFrame(perc_incr_contrib, columns=['perc_incr_contrib'])
    perc_contrib = pd.DataFrame(perc_contrib, columns=['perc_contrib'])
    roas = dataset.get_roas(plot=False)

    df = perc_incr_contrib.merge(perc_contrib, left_index=True, right_index=True, how = 'outer').merge(roas, left_index=True, right_index=True, how = 'outer')
    df.index = df.index.map(lambda x: rename_dict[x])
    
    weekly_df = dataset.get_media_contribution().loc[:,:'y_baseline']
    weekly_df = weekly_df.melt(ignore_index=False, var_name='channel', value_name='contribution')
    weekly_df['channel'] = weekly_df['channel'].map(lambda x: rename_dict[x])
    
    return df, weekly_df
```

```python
df, weekly_df = load_dataset('fitted_dataset.pkl')
```

# Plots

```python
category_order = weekly_df.groupby('channel').sum().sort_values('contribution', ascending=False).index
px.area(data_frame=weekly_df.reset_index(), x='wk_strt_dt', y='contribution',
        category_orders={'channel' : list(category_order)},
        color='channel')
```

```python
px.bar(data_frame=df[['perc_incr_contrib']].dropna(),
       title='Media Contribution Percentage',
       category_orders={'index' : list(df.sort_values('perc_incr_contrib', ascending=False).index)},)
```
